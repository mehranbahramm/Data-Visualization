CLUSTERED BAR

```python
sales_by_region_Discount = data.groupby(['Region', 'Discount'])['Sales'].sum().reset_index()

plt.figure(figsize=(15, 8), dpi=200)
sns.barplot(x='Region', y='Sales', hue='Discount', data=sales_by_region_Discount,color='b')
plt.xticks(rotation=60)
plt.title('Sales by Discount and Region')
plt.show()
```

STACKED BAR

```python
sales_by_region_segment = data.groupby(['Region', 'Segment'])['Sales'].sum().unstack()
sales_by_region_segment.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Sales by Segment and Region')
plt.show()
```
